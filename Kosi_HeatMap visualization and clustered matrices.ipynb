```python
import numpy as np
from numpy.random import randn
import pandas as pd

# for stats 
from scipy import stats 

#plotting libraries
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

#inline viz
%matplotlib inline
```

```python
flight_dframe = sns.load_dataset("flights")
```

```python
flight_dframe.head()
```

```python
flight_dframe = flight_dframe.pivot(index = "month", columns= "year",values= "passengers") 

flight_dframe
```

```python
sns.heatmap(flight_dframe, alpha = 0.9)
```

```python
sns.heatmap(flight_dframe, alpha = 0.9, annot = True, fmt = 'd')
```

```python
sns.heatmap(flight_dframe , center =  flight_dframe.loc['Aug', 1955])
```

```python
f, (axis1, axis2) = plt.subplots(2,1, figsize=(10,8))

yearly_flights = flight_dframe.sum()

years = pd.Series(yearly_flights.index.values)
years = pd.DataFrame(years)

flights = pd.Series(yearly_flights.values)
flights = pd.DataFrame(flights)

year_dframe = pd.concat((years,flights), axis =1)
year_dframe.columns = ['Year','Flights']

sns.barplot(x='Year', y = 'Flights', data = year_dframe, ax = axis1)
axis1.set_xticklabels(axis1.get_xticklabels(), rotation = 0)

sns.heatmap(flight_dframe, cmap= 'Blues', ax = axis2, cbar_kws = {'orientation':  'horizontal'}, vmin=0, vmax=flight_dframe.values.max())
```
