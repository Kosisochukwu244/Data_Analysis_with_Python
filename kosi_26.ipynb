```python
#first we'd import our usuals :
import pandas as pd
import numpy as np
from pandas import Series , DataFrame
from numpy.random import randn
```

```python
#next we import some useful data visual. tool:
from scipy import stats
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline 
```

```python
dset = randn(25)
```

```python
sns.rugplot(dset)

plt.ylim(0,1)
plt.hist(dset, color = 'indianred', alpha = 0.5)
```

```python
plt.hist(dset, color = 'indianred', alpha = 0.5)
```

```python
sns.rugplot(dset)
x_min = dset.min() - 2 
x_max = dset.max() + 2

x_axis = np.linspace(x_min, x_max, 100)

bandwidth = ((4 * dset.std() **5)/ 3* len(dset))** 0.2 

kernel_list = []

for dpoint in dset:
    #create a kernel for each point and append it to the kernel list
    kernel = stats.norm(dpoint, bandwidth).pdf(x_axis)
    kernel_list.append(kernel)


    #scale for plotting
    kernel = kernel/kernel.max()
    kernel = kernel * 0.4
    plt.plot(x_axis, kernel, color = 'grey', alpha = 0.3)
plt.ylim(0,1)
```

```python
sum_of_kde = np.sum(kernel_list,axis = 0)

fig = plt.plot(x_axis,sum_of_kde, color = 'indianred')

sns.rugplot(dset)
plt.yticks([])

plt.suptitle('sum of the basis function by kosi.py ')
```

```python
sns.kdeplot(dset)

plt.suptitle('KDE plot in one simple step by kosi.py')
```

```python
sns.rugplot(dset, color = 'black')

for bw in np.arange(0.5,2,0.25):
    sns.kdeplot(dset, bw = bw ,lw = 1.8, label = bw)
```
